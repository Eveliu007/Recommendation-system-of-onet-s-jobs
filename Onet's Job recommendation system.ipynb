```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df16=df15.fillna(0)
```

```python
sm = pd.DataFrame(cosine_similarity(df16))
sm.set_index(df16.index,inplace=True)
sm
```

```python
rm = pd.DataFrame()
for i in range(len(sm)):
    sm2 = sm.sort_values(by=i,ascending=False)
    rm[i] = sm2.index[0:20]
rm = rm.T
rm.set_index(df16.index, inplace=True)
```

```python
rm
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarities = linear_kernel(df16, df16)
```

```python
df16.dtypes
```

```python
df17=pd.read_csv('features(ID).csv')
df18=df17.fillna(0)
df18=df18.drop(['O*NET-SOC Code'],axis=1)
df18['ID'] = df18['ID'].astype(int)
```

```python
df18.dtypes
```

```python
results = {}

for idx, row in df18.iterrows():
    similar_indices = cosine_similarities[idx].argsort()[:-100:-1]
    similar_items = [(cosine_similarities[idx][i], df18['ID'][i]) for i in similar_indices]

    results[row['ID']] = similar_items[1:]

print('done!')
```

```python
results
```

```python
def item(id):
    return df18.loc[df18['ID'] == id]['ID'].values[0]
```

```python
def recommend(item_id, num):
    print("Recommending " + str(num) + " products similar to " + str(item(item_id)) + "...")
    print("-------")
    recs = results[item_id][:num]
    for rec in recs:
        print("Recommended: " + str(item(rec[1])) + " (score:" + str(rec[0]) + ")")
```

```python
recommend(item_id=1, num=5)
```
